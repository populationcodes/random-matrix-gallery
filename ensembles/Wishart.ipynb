```python
import plot
import utils

import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./presentation.mplstyle')
```

```python
N = 500
J = 1
P = 250

fig, s = plt.subplots(1, 1, figsize=(4, 4))

alpha = P/N
# Xij~Gaussian(0,J**2/N) and W=X*X'
X = np.random.randn(P, N)*J/math.sqrt(N)
W = X@X.T
z = np.linalg.eigvals(W) 

cx = s
plot.show_real_spectrum(cx, z, only_positive=True)
plot.text(cx, 0.6, 0.75, '$\\alpha=%1.1f$'%alpha)
lambda0 = J**2*(1-np.sqrt(alpha))**2
lambda1 = J**2*(1+np.sqrt(alpha))**2
x = np.linspace(lambda0, lambda1)
cx.plot(x, np.zeros_like(x), 'C1', linewidth=1.5, zorder=10, clip_on=False)
cx.plot(x, np.sqrt((lambda1-x)*(x-lambda0))/(2*np.pi*J**2*x*alpha), 'C0')
if alpha>1:
    cx.plot([0, 0], [0, 1-1/alpha], 'C0', zorder=10, clip_on=False, linewidth=2)
#fig.suptitle("Wishart ensemble (Marchenko–Pastur law)", fontsize="small")
name = utils.get_script_name()
if name is not None:
    fig.savefig("../images/%s.png"%name)
```

```python
N:int = 500
J = 1
Ps = [250, 500, 750, 1000]

fig, s = plt.subplots(1, len(Ps), figsize=(4*len(Ps), 4))
for i in range(len(Ps)):
    P = Ps[i]
    alpha = P/N
    # Xij~Gaussian(0,J**2/N) and W=X*X'
    X = np.random.randn(P, N)*J/math.sqrt(N)
    W = X@X.T
    z = np.linalg.eigvals(W) 
    
    cx = s[i]
    plot.show_real_spectrum(cx, z, only_positive=True)
    plot.text(cx, 0.6, 0.75, '$\\alpha=%1.1f$'%alpha)
    lambda0 = J**2*(1-np.sqrt(alpha))**2
    lambda1 = J**2*(1+np.sqrt(alpha))**2
    x = np.linspace(lambda0, lambda1)
    cx.plot(x, np.zeros_like(x), 'C1', linewidth=1.5, zorder=10, clip_on=False)
    cx.plot(x, np.sqrt((lambda1-x)*(x-lambda0))/(2*np.pi*J**2*x*alpha), 'C0')
    if alpha>1:
        cx.plot([0, 0], [0, 1-1/alpha], 'C0', zorder=10, clip_on=False, linewidth=2)

fig.suptitle("Wishart ensemble (Marchenko–Pastur law)", fontsize="small")
```
