```python
import plot
import utils

import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./presentation.mplstyle')
```

```python
utils.init_randomness()

N = 500
J = 2

# Wji~Gaussian(0,J**2/N)
W = np.random.randn(N,N)*J/math.sqrt(N)
z = np.linalg.eigvals(W)
m = 1.1*np.max(np.abs(z))

fig, s = plt.subplots(1, 1, figsize=(4,4))
cx = s
plot.show_complex_spectrum(cx, z, radius=[J,J])

name = utils.get_script_name()
if name is not None:
    fig.savefig("../images/%s.png"%name)
plt.show()
```

```python
N = 500
Js = [0.5, 1, 2]

fig, s = plt.subplots(1, len(Js), figsize=(4*len(Js),4))
for i in range(len(Js)):
    J = Js[i]
    # Wji~Gaussian(0,J**2/N)
    W = np.random.randn(N,N)*J/math.sqrt(N)
    z = np.linalg.eigvals(W)
    m = 1.1*np.max(np.abs(z))

    cx = s[i]
    plot.show_complex_spectrum(cx, z, radius=[J,J])
    plot.text(cx, 0.8, 0.8, '$J=%1.1f$'%J, fontsize='xx-small')

fig.suptitle("Ginibre Ensemble => Circular law", fontsize="small")
plt.show()
```
